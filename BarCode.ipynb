```python
L_Group = ["LLLLLL", "LGGLLG", "LLGLGG", "LGGGLL", "LLGGLG", "LGLGLG", "LLGGGL", 
           "LGLGGL", "LGLLGG", "LGGLGL"]

R_Group = ["RRRRRR","RRRRRR", "RRRRRR","RRRRRR", "RRRRRR","RRRRRR", "RRRRRR",
           "RRRRRR", "RRRRRR","RRRRRR"]

L_Code = ["WWWBBWB", "WWBBWWB", "WWBWWBB", "WBBBBWB", "WBWWWBB", "WBBWWWB", 
          "WBWBBBB", "WBBBWBB", "WBBWBBB", "WWWBWBB"]

G_Code = ["WBWWBBB", "WBBWWBB", "WWBBWBB", "WBWWWWB", "WWBBBWB", "WBBBWWB", 
          "WWWWBWB", "WWBWWWB", "WWWBWWB", "WWBWBBB"]

R_Code = ["BBBWWBW", "BBWWBBW", "BBWBBWW", "BWWWWBW", "BWBBBWW", "BWWBBBW", 
          "BWBWWWW", "BWWWBWW", "BWWBWWW", "BBBWBWW"]


import Graphics

# 0 Put your name in a comment at the top of the program.

# 1. Ask the user for a 13-character string, all digits. If the string is shorter than 13 characters, pad it
# on the left with 0 characters to make it exactly 13 digits. Convert the string into a list of 13
# integers, where each integer is in the range [0...9].
user = input("Enetr 13 digits number: ")
if len(user)<13:
    user = '0'+'-'+user[:6]+'-'+user[6:]
if len(user)==13:
    user = user[:1]+'-'+user[1:7]+'-'+user[7:]

# 2. Ask the user for a scale (1-10). This represents the width of a stripe, in pixels (that is, at
# Scale=1 a stripe is one pixel wide; at Scale=5 a stripe is five pixels wide).
scale_factor = int(input("Enter scale(1-10)"))



# 3. Determine the width and height of the canvas, based on the number of needed stripes and the
# Scale value. Similarly, determine the length of the long and short lines. Create the canvas.
Canvas = Graphics.makeEmptyPicture(scale_factor*113, scale_factor*113, Graphics.white)

length_long_bar = (113 - 95) * scale_factor
length_short_bar = (113- 90) * scale_factor

# 4. Based on the first of the 13 digits, determine from L_Group and R_Group the code patterns
# to use.
lgroup = L_Group[int(user[0])]   # 
rgroup = R_Group[int(user[0])]   # 

S = ((scale_factor*113)-(95*scale_factor))/2  # customize distace from left and right corner
T = ((113 - 95) * scale_factor)/2             # customize distance from top

# 5. Paint the left guard pattern with long lines.
for y,z in zip(range(0, 3*scale_factor, scale_factor), (list("BWB"))):
    if z == 'W': continue
    else:
        for X in range(y,y+scale_factor):
            Graphics.VerticalLine(Canvas, X+S, 50, Graphics.getHeight(Canvas)-length_long_bar, Graphics.black)


# 6. Paint digits 2-7 as a series of black and white stripes with short lines, with the patterns picked
# from L_Code and G_Code
bw_lft_mdl = ''
for a,b in zip(range(2,8), range(0,7)):
    if lgroup[b]=='L':
        bw_lft_mdl += L_Code[int(user[a])]
    if lgroup[b]=='G':
        bw_lft_mdl += G_Code[int(user[a])]
    if lgroup[b]=='R':
        bw_lft_mdl += R_Code[int(user[a])]
        
for y,z in zip(range(3*scale_factor, 45*scale_factor, scale_factor), (list(bw_lft_mdl))):
    if z == 'W': continue
    else:
        for X in range(y,y+scale_factor):
            Graphics.VerticalLine(Canvas, X+S, 50, Graphics.getHeight(Canvas)-length_short_bar, Graphics.black)
    
# 7. Paint the middle guard pattern with long lines.
#Mid guard bar WBWBW
for y,z in zip(range(45*scale_factor, 50*scale_factor, scale_factor), (list("WBWBW"))):
    if z == 'W': continue
    else:
        for X in range(y,y+scale_factor):
            Graphics.VerticalLine(Canvas, X+S, 50, Graphics.getHeight(Canvas)-length_long_bar, Graphics.black)



# 8. Paint digits 8-13 as a series of black and white stripes with short lines, with the patterns picked
# from R_Code.

bw_mdl_rgt = ''
for a,b in zip(range(9,15), range(0,7)):
    if rgroup[b]=='L':
        bw_mdl_rgt += L_Code[int(user[a])]
    if rgroup[b]=='G':
        bw_mdl_rgt += G_Code[int(user[a])]
    if rgroup[b]=='R':
        bw_mdl_rgt += R_Code[int(user[a])]
        
for y,z in zip(range(50*scale_factor, 92*scale_factor, scale_factor), (list(bw_mdl_rgt))):
    if z == 'W': continue
    else:
        for X in range(y,y+scale_factor):
            Graphics.VerticalLine(Canvas, X+S, 50, Graphics.getHeight(Canvas)-length_short_bar, Graphics.black)


# 9. Paint the right guard pattern with long lines.
for y,z in zip(range(92*scale_factor, 95*scale_factor, scale_factor), (list("BWB"))):
    if z == 'W': continue
    else:
        for X in range(y,y+scale_factor):
            Graphics.VerticalLine(Canvas, X+S, 50, Graphics.getHeight(Canvas)-length_long_bar, Graphics.black)

# 10. Save the image as BarCode.bmp on your disk.
Graphics.WriteBMP("BarCode", Canvas)
```
